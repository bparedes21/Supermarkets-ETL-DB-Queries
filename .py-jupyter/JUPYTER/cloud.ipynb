```python
from wordcloud import WordCloud, STOPWORDS
```

```python
#archivo para crear  lista de sucursal_id
#nombre del archivo sucursal
url = "https://drive.google.com/file/d/1skuhvdTkgh_GZ9wL0FEf3uxSZu7wU-mX/view?usp=sharing"
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
#archivo csv que descargo desde el drive
Archivo_sucursal= pd.read_csv(dwn_url)
```

```python
sucursal_list=list(Archivo_sucursal["banderaDescripcion"])
```

```python
string=""
for i in sucursal_list:
    string=string+", "+i
```

```python
wordcloud = WordCloud(width=1600, height=800).generate(string)
# Open a plot of the generated image.

plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)

plt.show()

wordcloud.to_file("cloud_super.png")
```
